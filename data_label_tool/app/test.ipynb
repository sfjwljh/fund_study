```python
import json
with open('/Users/liujianhui02/Desktop/sync/obsidian/Master/fund_stream_project/codes/data_label_tool/app/alpaca_train.json','r',encoding='utf-8') as f:
    doc=json.loads(f.read())
```

```python
doc
```

```python
len_dis=[]
for en in doc:
    len_dis.append(len(en['input']))
len_dis=sorted(len_dis)
```

```python
import statistics

data = len_dis

mean_val = statistics.mean(data)       # 均值
median_val = statistics.median(data)   # 中位数
mode_val = statistics.mode(data)       # 众数（注意：如果有多个众数，statistics.mode 只返回第一个）
variance_val = statistics.variance(data) # 方差
stdev_val = statistics.stdev(data)       # 标准差
min_val = min(data)                    # 最小值
max_val = max(data)                    # 最大值

print("使用 statistics 模块计算的描述统计：")
print(f"均值: {mean_val}")
print(f"中位数: {median_val}")
print(f"众数: {mode_val}")
print(f"方差: {variance_val}")
print(f"标准差: {stdev_val}")
print(f"最小值: {min_val}")
print(f"最大值: {max_val}")
```
