```python
sql="select code from total"
cursor.execute(sql)
db_code_list=cursor.fetchall()
db_code_list2=[code[0] for code in db_code_list]
```

```python

# 把txt里的条目都加到数据库
for code in txt_list:
    if code in db_code_list2:
        update_query = "UPDATE total SET stt = 1 WHERE code = %s"
        cursor.execute(update_query, (code))
        print('更新了{}stt=1'.format(code)) 
    else:
        #新增一行
        insert_query = "INSERT INTO total (code,stt) VALUES (%s, %s)"
        cursor.execute(insert_query, (code, 1))
        print('新增了{}'.format(code))
db.commit()
```

```python
# 更新数据库的stt
for code in db_code_list2:
    check_query = "SELECT stt FROM total WHERE code = %s"
    cursor.execute(check_query, (code))
    result=cursor.fetchone()
    if code in txt_list and result[0]==1:
        continue
    elif code not in txt_list and result[0]==0:
        continue
    elif code in txt_list and result[0]==0:
        update_query = "UPDATE total SET stt = 1 WHERE code = %s"
        cursor.execute(update_query, (code))
        print('更新了{}stt=1'.format(code))
    elif code not in txt_list and result[0]==1:
        update_query = "UPDATE total SET stt = 0 WHERE code = %s"
        cursor.execute(update_query, (code))
        print('更新了{}stt=0'.format(code))
db.commit() 
```
