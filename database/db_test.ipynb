# 第一次把pickle所有文件导入到数据库

```python
# 批量插入新数据行
for i in range(0,len(content)):
    line=content.loc[i]
    # 要插入的数据如下
    new_data = {
        'code': line['code'],
        'date': line['date'],
        'title': line['title'],
        'company':line['company'],
        'host':line['host'],
        'm3u8_url':line['catch_url'],


    }

    # 检查主键code是否存在
    check_query = "SELECT code FROM total WHERE code = %s"
    cursor.execute(check_query, (new_data['code'],))
    result = cursor.fetchone()

    if result:
        # print("ID already exists, cannot insert.")
        pass
    else:
        # 插入数据
        insert_query = "INSERT INTO total (code, date, title,company,host,m3u8_url) VALUES (%s, %s, %s, %s, %s, %s)"
        cursor.execute(insert_query, (new_data['code'], new_data['date'], new_data['title'], new_data['company'], new_data['host'], new_data['m3u8_url']))
        db.commit()
        # print("Data inserted successfully.")

# # 关闭连接
# cursor.close()
# db.close()
```

# 把picke的catch_url导入到数据库的m3u8_url列中

```python
# 插入新列
for i in range(0,len(content)):
    print("dealing with{}".format(i))
    line=content.loc[i]
    if len(line['catch_url'])==0:
        print('len=0,pass')
        continue
    # 要插入的数据如下
    new_data = {
        'code': line['code'],
        'm3u8_url':line['catch_url'],
    }
    check_query = "SELECT code,m3u8_url FROM total WHERE code = %s"
    cursor.execute(check_query, (new_data['code']))
    result = cursor.fetchone()

    if result!=None:
        # 更新数据
        if result[1]==new_data['m3u8_url']:
            print('already exist,pass')
            continue
        update_query = "UPDATE total SET m3u8_url =%s  WHERE CODE = %s"
        cursor.execute(update_query, (new_data['m3u8_url'], new_data['code']))
        db.commit()
        print('updated successfully')
    else:
        print("{}该行不存在".format(line['code']))
```
